```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, grade in enumerate(['G1', 'G2', 'G3']):
    sns.histplot(df_mat[grade], bins=20, kde=True, ax=axes[i], color=['#3498db', '#9b59b6', '#e67e22'][i])
    axes[i].set_title(f'{grade} Distribution')
    axes[i].set_xlabel('Grade (0-20)')
    axes[i].set_ylabel('Count')

plt.suptitle('Grade Distributions Across All Periods', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Create binary target: 1 = at-risk (failing), 0 = passing

df_mat['at_risk'] = (df_mat['G3'] < 10).astype(int)

# Counts
risk_counts = df_mat['at_risk'].value_counts().sort_index()
total = len(df_mat)

# Plot
plt.figure(figsize=(6,4))
ax = sns.countplot(x='at_risk', data=df_mat,hue='at_risk',  palette=['green', 'red'],legend = False)

# Fix labels
ax.set_xticklabels(['Passing', 'At-Risk'])
ax.set_title('Count of Students by Risk Status')
ax.set_xlabel('')

for p in ax.patches:
    count = int(p.get_height())
    percentage = 100 * count / total
    ax.annotate(f'{count} ({percentage:.1f}%)',
                (p.get_x() + p.get_width() / 2., count),
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Absences vs Risk
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(x='at_risk', y='absences', data=df_mat,
            palette=['#2ecc71', '#e74c3c'], hue = "at_risk", ax=axes[0], legend=False)
axes[0].set_xticklabels(['Passing', 'At-Risk'])
axes[0].set_title('Absences vs Risk Status')
axes[0].set_xlabel('')

# Study Time vs Risk
study_risk = df_mat.groupby(['studytime', 'at_risk']).size().unstack()
study_risk.plot(kind='bar', ax=axes[1], color=['#2ecc71', '#e74c3c'], edgecolor='white')
axes[1].set_xticklabels(['<2 hrs', '2-5 hrs', '5-10 hrs', '>10 hrs'], rotation=0)
axes[1].set_title('Study Time vs Risk Status')
axes[1].legend(['Passing', 'At-Risk'])
axes[1].set_xlabel('Weekly Study Time')

plt.tight_layout()
plt.show()
```

```python
# Past Failures vs Risk
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

failures_risk = df_mat.groupby(['failures', 'at_risk']).size().unstack(fill_value=0)
failures_risk.plot(kind='bar', ax=axes[0], color=['#2ecc71', '#e74c3c'], edgecolor='white')
axes[0].set_xticklabels(['0', '1', '2', '3+'], rotation=0)
axes[0].set_title('Past Failures vs Risk Status')
axes[0].legend(['Passing', 'At-Risk'])
axes[0].set_xlabel('Number of Past Failures')

# Alcohol Consumption vs G3
sns.scatterplot(x='Dalc', y='G3', hue='at_risk', data=df_mat,
                palette=['green', 'red'], ax=axes[1])
axes[1].axhline(y=10, color='black', linestyle='--')
axes[1].set_title('Weekday Alcohol Consumption vs Final Grade')
axes[1].set_xlabel('Alcohol Consumption (1=Low, 5=High)')
axes[1].legend( labels=['At-Risk', 'Passing'])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Gender vs Risk
sex_risk = df_mat.groupby(['sex', 'at_risk']).size().unstack()
sex_risk.plot(kind='bar', ax=axes[0], color=['#2ecc71', '#e74c3c'], edgecolor='white', rot=0)
axes[0].set_xticklabels(['Female', 'Male'])
axes[0].set_title('Gender vs Risk')
axes[0].legend(['Passing', 'At-Risk'])
axes[0].set_xlabel('')

# Age Distribution
sns.histplot(data=df_mat, x='age', hue='at_risk', multiple='dodge',
             palette=['green', 'red'], ax=axes[1])
axes[1].set_title('Age Distribution by Risk')
axes[1].legend(title='At Risk', labels=['At-Risk', 'Passing'])

# Address (Urban vs Rural) vs Risk
addr_risk = df_mat.groupby(['address', 'at_risk']).size().unstack()
addr_risk.plot(kind='bar', ax=axes[2], color=['#2ecc71', '#e74c3c'], edgecolor='white', rot=0)
axes[2].set_xticklabels(['Rural', 'Urban'])
axes[2].set_title('Address Type vs Risk')
axes[2].legend(['Passing', 'At-Risk'])
axes[2].set_xlabel('')

plt.suptitle('Demographic Analysis', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Mother's Education vs G3
sns.boxplot(x='Medu', y='G3', data=df_mat, hue = "Medu", palette='Blues', legend = False, ax=axes[0])
axes[0].axhline(y=10, color='red', linestyle='--', alpha=0.7)
axes[0].set_title("Mother's Education vs Grade")
axes[0].set_xlabel('Education Level (0=None, 4=Higher)')

sns.boxplot(x='Fedu', y='G3', data=df_mat, hue = "Fedu", palette='Greens', legend = False, ax=axes[1])
axes[1].axhline(y=10, color='red', linestyle='--', alpha=0.7)
axes[1].set_title("Father's Education vs Grade")
axes[1].set_xlabel('Education Level (0=None, 4=Higher)')


plt.suptitle('Family Factors', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
# Family Relationship Quality vs G3
sns.boxplot(x='famrel', y='G3', data=df_mat, hue = "famrel", palette='Greens', legend = False, ax=axes[0])
axes[0].axhline(y=10, color='red', linestyle='--', alpha=0.7)
axes[0].set_title('Family Relationship Quality vs Grade')
axes[0].set_xlabel('Family Rel. (1=Very Bad, 5=Excellent)')

# Going Out vs G3
sns.boxplot(x='goout', y='G3', data=df_mat, hue = "goout", legend = False, palette='Oranges', ax=axes[1])
axes[1].axhline(y=10, color='red', linestyle='--', alpha=0.7)
axes[1].set_title('Going Out Frequency vs Grade')
axes[1].set_xlabel('Going Out (1=Very Low, 5=Very High)')


plt.suptitle('Social Factors', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Select numeric columns only
numeric_cols = df_mat.select_dtypes(include=np.number).columns.tolist()

plt.figure(figsize=(14, 10))
corr_matrix = df_mat[numeric_cols].corr()

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # Hide upper triangle
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
            cmap='RdYlGn', center=0, linewidths=0.5,
            annot_kws={'size': 7})

plt.title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Correlation with G3
g3_corr = corr_matrix['G3'].drop('G3').sort_values(key=abs, ascending=False)

colors = ['#2ecc71' if v > 0 else '#e74c3c' for v in g3_corr]

plt.figure(figsize=(10, 7))
g3_corr.plot(kind='barh', color=colors, edgecolor='white')
plt.axvline(x=0, color='black', linewidth=0.8)
plt.title('Feature Correlations with Final Grade (G3)', fontsize=13, fontweight='bold')
plt.xlabel('Correlation Coefficient')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

support_features = ['schoolsup', 'famsup', 'internet']
titles = ['School Support', 'Family Support', 'Internet Access']

for i, (feat, title) in enumerate(zip(support_features, titles)):
    feat_risk = df_mat.groupby([feat, 'at_risk']).size().unstack(fill_value=0)
    feat_risk_pct = feat_risk.div(feat_risk.sum(axis=1), axis=0) * 100
    feat_risk_pct.plot(kind='bar', ax=axes[i], color=['#2ecc71', '#e74c3c'],
                       edgecolor='white', rot=0)
    axes[i].set_title(f'{title} vs Risk (%)')
    axes[i].legend(['Passing', 'At-Risk'])
    axes[i].set_ylabel('Percentage (%)')
    axes[i].set_xlabel('')

plt.suptitle('Support System', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
